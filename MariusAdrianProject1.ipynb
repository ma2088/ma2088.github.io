```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns



# Load the dataset 
df = pd.read_csv('Dataset2/shanghai_ranking_2024.csv')

# Display the first 10 rows in a neatly formatted table
table = tabulate(df.head(20), headers='keys', tablefmt='github', showindex=False)
print(table)

```

Sub-questions
1.) What is the Relationship between 'Award' and 'PCP'?
2.) What is the relationship between the highly cited researchers score (Hici) and PCP?
3.) What is the Distribution of PCP with using a Violin Plot?



```python
#Sub-question: Relationship between 'Award' and 'PCP'
plt.figure(figsize=(8,6))
sns.scatterplot(x='Award', y='PCP', data=df)
plt.title('Award Score vs Per Capita Performance')
plt.xlabel('Award Score')
plt.ylabel('PCP (Per Capita Performance)')
plt.show()
```

```python
#Relationship between the highly cited researchers score (Hici) and PCP?
plt.figure(figsize=(10,6))
sns.scatterplot(x='Hici', y='PCP', data=df)
plt.title('Highly Cited Researchers (Hici) vs Per Capita Performance (PCP)')
plt.xlabel('Highly Cited Researchers Score (Hici)')
plt.ylabel('Per Capita Performance (PCP)')
plt.show()
```

```python
# Sub-question: Distribution of Per Capita Performance (PCP) using a Violin Plot
plt.figure(figsize=(8,6))
sns.violinplot(y='PCP', data=df)
plt.title('Distribution of Per Capita Performance (PCP)')
plt.ylabel('PCP Score')
plt.show()
```

Summary of Findings:

Award Scores vs PCP:

In the scatterplot comparing Award Scores with PCP, we notice a moderate correlation. While some universities with high award scores also have high PCP scores, the correlation is not as strong as with research metrics (Hici, PUB). This indicates that alumni prestige and faculty awards contribute to per capita performance, but not as significantly as research output and citations.


Highly Cited Researchers (Hici) vs PCP:

The scatterplot for Highly Cited Researchers (Hici) vs PCP reveals a strong positive relationship, indicating that universities with more highly cited researchers tend to have higher PCP scores. This suggests that universities with a high concentration of influential researchers are more efficient per capita in their performance.

Distribution of Per Capita Performance (PCP)

The distribution and density of PCP values among colleges are depicted in vivid detail by the violin plot. You may determine if there is much variety or if most colleges are clustered around certain performance levels based on the plot's shape, skewness, and concentration. For example, you may conclude that a few highly performing colleges are responsible for the high levels of per capita performance if there is a long tail of high PCP scores. On the other hand, a close distribution surrounding a median would indicate greater consistency in PCP amongst institutions.
