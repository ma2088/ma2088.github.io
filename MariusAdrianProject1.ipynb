Summary of Findings:

Essential Elements Influencing University Rankings:
Based on the investigation, it was shown that a university's global ranking is strongly correlated with specific measures, including Award ratings and Hici (Highly Cited Researchers). Higher-ranking universities worldwide often had higher scores in these categories.
Rankings were also heavily influenced by alum scores, with elite universities typically having higher alum scores due to their illustrious academic and historical backgrounds.

Variations in Per Capital Performance (PCP):
Per Capital Performance (PCP) showed a clear difference between universities in the top 100 and those placed lower. In general, the PCP ratings of the top 100 universities were much higher, indicating that these universities are more productive in producing high-impact research and academic output than their size.

Publications and Nature & Science articles: 
The overall number of publications (PUB) and the number of articles in Nature & Science (N&S) were positively correlated. High-volume publishing universities also often did an excellent job of publishing their best work in esteemed journals.
   
Regional Comparisons: 
Universities in more competitive regions (i.e., nations with highly ranked universities) earned higher scores across various measures, including publications and Nature & Science papers, according to a study of scores based on inferred regions (using national/regional rank).



Gaps and Difficulties: 

Missing Regional Information: It was challenging to do an accurate regional comparison because the dataset lacked a clear region or nation column. For a more detailed analysis, more information about specific nations or regions would be helpful.
Outliers and Data Distribution: Although there were outliers in several variables, such as Award and Alumni scores, deleting them did not substantially change the overall patterns that were seen. Nonetheless, addressing outliers more skillfully could enhance insights.
Data on Nature and Science (N&S): The N&S metric has missing values in multiple rows. The mean was used to fill in these missing values, however this method might have decreased the validity of those particular analyses.
